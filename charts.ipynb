```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = [1,2,4,6,7]
y = [10,20,30,40,50]
```

```python
plt.plot(x,y,c='m',label='simple line plot') 
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('line plot magenta')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y,label='scatter plot',c="g",marker="*")
plt.xlabel('x values')
plt.ylabel('y values')

plt.title('scatter plot magenta')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y ,c='g',label="scatter plot")
plt.plot(x,y,c='m',label='line plot')
plt.xlabel('x values')
plt.ylabel("y values")
plt.title('line with scatter plot')
plt.legend()
plt.show()
```

```python
x=['A','B','C']
y=[20,30,40]
```

```python
plt.bar(x,y,label="bar graph",color='b')
plt.title("Bar ")
plt.xlabel('x values')
plt.ylabel("y values")
plt.show()
```

```python
label=['A','B','C']
x=[20,10,40]
color=['g','m','y']
explode=[0,0.1,0]
```

```python
plt.pie(x,labels=label,colors=color,autopct='%5.1f%%',explode=explode,shadow=True)
plt.show()
```
