# jointplot

A joinplot allows to study the realationship b/w 2 numerical variable. The central chart display their correlation.
It is usally a scatter plot,a hexbin plot ,a 2D histogram or 2D density plot.
this comes under the univariate analysis.

```python
sns.jointplot(x='tip',y="total_bill",data=df, kind='hex')
```

```python
sns.jointplot(x='tip',y="total_bill",data=df, kind='reg',color='g')
```

# pair plot

If the data set has a more than two features that time use this plot.
   pair plot also known as scatter plot.In which one variable in the same data rows is matched with another variable values,
like this: pair polts are just elaborations on this showing all variable paired with all the other variables.
Pair plot do the  permitations and combination of all the features.
for pair plot features should integer and float columns.


```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

All garaphs plotted based on the sex columns

# Dist plot

Dist plot helps us to create the histogram.
Dist plot help us to check the distribution of the columns features.


```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

kde means kernal density estimation.

It shows the counts of observations in each categorical bon using bars.

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

# Bar plot

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

# Boxp plot

Box plot also called as whisker plot.
This graph presents information from five-number summary.


```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v') #this did box plot for whole data set. 
```

box plot done for only numericcal columns

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
print("box plot based on the smoker feature")
```

# Violin plot

It help us to see 'the distribution of data in terms of kde and the barplot.
It is combination of kde and box plot.

```python
sns.violinplot(x='day',y='total_bill',data=df,palette="rainbow")
```
